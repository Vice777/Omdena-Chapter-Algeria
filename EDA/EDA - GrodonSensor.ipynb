```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = ("../Dataset")
os.listdir(path)
```

```python
files = ['AICU', 'Automatoes', 'Digilog', 'IUACAAS', 'Reference', 'TheAutomators']
```

## Grodan Sensors

```python
df = {}
for file in files:
    df[file] = pd.read_csv("../Dataset/"+file+"/GrodanSens.csv")
```

```python
for i in df:
    print(i,"\n",df[i].head(),"\n")
    
```

```python
for i in df:
    df[i]['%time'] = pd.to_datetime(df[i]['%time'], unit='d',origin='1899-12-30')
```

```python
col = df['AICU'].columns
col = col[1:]
```

```python
#as the given data is in time-series format therefor interpolating the data with nearest member to remove null values

for i in df:
    for j in col:
        df[i][j] = df[i][j].astype(str).astype(float)

for i in df:
    df[i] = df[i].fillna(df[i].mean())  
  
```

```python
for i in df:
    print(i,"\n",df[i].head(),"\n")
```

```python
from pandas.plotting import scatter_matrix
df_corr = df['AICU'].corr()

f, ax = plt.subplots(figsize=(6, 6))
sns.heatmap(df_corr, mask=np.zeros_like(df_corr, dtype=np.bool), cmap = "BrBG",ax=ax)
plt.show()
```

```python

def plot(target):
    
    sns.set()
    plt.figure(figsize=(20, 20),facecolor='lightgray')

    plt.subplot(321)
    for i in df:
        plt.plot(df[i][target], df[i]["EC_slab1"])
    plt.legend(files)
    plt.title("EC_slab1 vs {target}".format(target=target))
    plt.xlabel('EC_slab1')
    plt.ylabel(target)


    plt.subplot(322)
    for i in df:
        plt.plot(df[i][target], df[i]['EC_slab2'])
    plt.legend(files)
    plt.title("EC_slab2 vs {target}".format(target=target))
    plt.xlabel('EC_slab2')
    plt.ylabel(target)


    plt.subplot(323)
    for i in df:
        plt.plot(df[i][target], df[i]['WC_slab1'])
    plt.legend(files)
    plt.title("WC_slab1 vs {target}".format(target=target))
    plt.xlabel('WC_slab1')
    plt.ylabel(target)


    plt.subplot(324)
    for i in df:
        plt.plot(df[i][target], df[i]['WC_slab2'])
    plt.legend(files)
    plt.title("WC_slab2 vs {target}".format( target=target))
    plt.xlabel('WC_slab2')
    plt.ylabel(target)

    plt.subplot(325)
    for i in df:
        plt.plot(df[i][target], df[i]['t_slab1'])
    plt.legend(files)
    plt.title("t_slab1 vs {target}".format( target=target))
    plt.xlabel('t_slab1')
    plt.ylabel(target)

    plt.subplot(326)
    for i in df:
        plt.plot(df[i][target], df[i]['t_slab2'])
    plt.legend(files)
    plt.title("t_slab2 vs {target}".format( target=target))
    plt.xlabel('t_slab2')
    plt.ylabel(target)
    
    plt.show()



#Stem_thick  Cum_trusses  stem_dens   plant_dens
```

```python
plot("%time")
```
