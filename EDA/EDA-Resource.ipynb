```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = ("../Dataset")
os.listdir(path)
```

```python
files = ['AICU', 'Automatoes', 'Digilog', 'IUACAAS', 'Reference', 'TheAutomators']
```

## Resources

```python
df = {}
for file in files:
    df[file] = pd.read_csv("../Dataset/"+file+"/Resources.csv")
```

```python
for i in df:
    df[i].info()
    print("")
```

```python
for i in df:
    print(i,"\n",df[i].head(),"\n")
    
```

```python
for i in df:
    df[i]['%Time '] = pd.to_datetime(df[i]['%Time '], unit='d',origin='1899-12-30')
```

```python
col = df['AICU'].columns
col = col[1:]
```

```python
for i in df:
    print(df[i].isnull().sum())
    print("")
```

```python
#as the given data is in time-series format therefor interpolating the data with nearest member to remove null values

for i in df:
    for j in col:
        df[i][j] = df[i][j].astype(str).astype(float)

for i in df:
    df[i] = df[i].fillna(df[i].mean())  
  
```

```python
for i in df:
    df[i].info()
    print("")
```

```python
for i in df:
    print(i,"\n",df[i].head(),"\n")
```

```python
for i in df:
    print(i,"\n",df[i].describe(),"\n")
```

```python
from pandas.plotting import scatter_matrix
k=0
plt.figure(figsize=(8,40))
for i in df:
    df_corr = df[i].corr()
    plt.subplot(6, 1, k + 1)
    g = sns.heatmap(df_corr, mask=np.zeros_like(df_corr, dtype=np.bool), cmap = "BrBG")
    plt.title(i)
    k=k+1
plt.show()
```

```python
plt.figure(figsize=(6,6))
for j in df:
    g = sns.pairplot(df[j])
    g.fig.suptitle("{}".format(j), y=1.08) # y= some height>1

plt.show()
```

```python
for i in df:
    plt.figure(figsize=(10,6))
    sns.displot(df[i],bins=5)
```

```python
def plot(target):
    
    k=0
    sns.set()
    plt.figure(figsize=(20, 60),facecolor='lightgray')

    for j in col:
        plt.subplot(6, 1, k + 1)
        for i in df:
            plt.plot(df[i][target], df[i][j])
        plt.legend(files)
        plt.title("{var} vs {target}".format(var=j,target=target))
        plt.xlabel(j)
        plt.ylabel(target)
        k=k+1
    plt.show()
        
    
```

```python
plot("%Time ")
```
